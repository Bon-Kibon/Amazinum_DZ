# Question 1

```python
import pandas as pd
import re #Для фільтрації рядка

#Energy
energy = pd.read_excel("Energy Indicators.xls", skiprows=16, skipfooter=38, engine="xlrd")
energy = energy.drop(energy.columns[:2], axis=1)
energy.columns = ["Country", "Energy Supply", "Energy Supply per Capita", "% Renewable"]

energy.loc[1: ,"Energy Supply"] *= 1000000
energy.loc[0,"Energy Supply"] = "Gigajoules"

energy = energy.map(lambda x: None if isinstance(x, str) and "..." in x else x)

#Видалення чисел
energy["Country"] = energy["Country"].apply(lambda x: re.sub(r'\d+', '', str(x)) )
#Видалення дужок
energy["Country"] = energy["Country"].apply(lambda x: re.sub(r'\(.*?\)', '', str(x)) )
#Видалення зайвих пробілів
energy['Country'] = energy['Country'].str.strip()

energy["Country"] = energy["Country"].replace({"Republic of Korea": "South Korea",
"United States of America": 'United States',
'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
'China, Hong Kong Special Administrative Region': 'Hong Kong'})

#GDP
gdp = pd.read_csv("API_NY.GDP.MKTP.CD_DS2_en_csv_v2_76261.csv", skiprows=3)
gdp["Country Name"] = gdp["Country Name"].replace({"Korea, Rep.": "South Korea",
"Iran, Islamic Rep.": "Iran",
"Hong Kong SAR, China": "Hong Kong"})
gdp = gdp.rename(columns={"Country Name" : "Country"})

#ScimEn
ScimEn = pd.read_excel("scimagojr country rank 1996-2023.xlsx")

def answer_one():
    df_1 = gdp.drop(columns=['1960',
       '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969',
       '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978',
       '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987',
       '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996',
       '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005'])
    df_2 = ScimEn.head(15)
    result = energy.merge(df_1, how='inner', on="Country").merge(df_2, how='inner',on="Country")
    result = result.set_index("Country")
    result = result[['Rank', 'Documents', 'Citable documents',
                    'Citations', 'Self-citations', 'Citations per document',
                    'H index', 'Energy Supply', 'Energy Supply per Capita',
                    '% Renewable', '2006', '2007', '2008', '2009', '2010',
                    '2011', '2012', '2013', '2014', '2015']]
    return(result)
    
answer_one()
```

# Question 2

```python
def answer_two():
    avgGDP = answer_one()[['2006', '2007', '2008', '2009', '2010','2011', '2012', '2013', '2014', '2015']].mean(axis=1)
    avgGDP = avgGDP.sort_values(ascending=False)
    return avgGDP
answer_two()
```

# Question 3

```python
def answer_three():
    tmp = answer_one().sort_values('Rank')
    tmp = tmp.set_index('Rank')
    result = tmp.at[6, '2015'] - tmp.at[6, '2006']
    if result < 0: #На всякий случай
        result = abs(result)
    return result

print(answer_three())
```

# Question 4

```python
def answer_four():
    tmp = answer_one()
    tmp['Ratio'] = tmp.loc[ :,'Self-citations'] / tmp.loc[:, 'Citations']
    max_ratio = tmp['Ratio'].max()
    county = tmp['Ratio'].idxmax()
    result = (county, max_ratio)
    return result

print(answer_four())
```

# Question 5

```python
def answer_five():
    tmp = answer_one()
    tmp['Population'] = tmp.loc[ :,'Energy Supply'] / tmp.loc[:, 'Energy Supply per Capita']
    tmp = tmp.sort_values('Population', ascending=False)
    result = tmp.index[2]
    return result

print(answer_five())
```

# Question 6

```python
def answer_six():
    tmp = answer_one()
    tmp['citable documents per person'] = tmp.loc[ :,'Population'] / tmp.loc[:, 'Citable documents']
    correlation = tmp['citable documents per person'].corr(tmp['Energy Supply per Capita'])
    return correlation
print(answer_six())
```

# Question 7

```python
def answer_seven():
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    tmp = answer_one()
    tmp['Population'] = tmp.loc[ :,'Energy Supply'] / tmp.loc[:, 'Energy Supply per Capita']
    tmp['Continent'] = ContinentDict
    df = tmp.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])   
    return(df)

answer_seven()
```
