```python
import pandas as pd
import numpy as np

df = pd.read_csv('./Groceries_dataset.csv')
df.head()
```

```python
products_df = pd.get_dummies(df['itemDescription'])
products_df.head()
```

```python
data = df.copy()
data.drop(['itemDescription'], axis = 1, inplace = True)
data = data.join(products_df)
data.head()
```

```python
products = df['itemDescription'].unique()
data = data.groupby(['Member_number', 'Date'])[products].sum()
data
```

```python
for col in products:
    data[col] = data[col].apply(lambda x: col if x != 0 else 0)
data
```

```python
newdata = data.values
newdata
```

```python
filtered_data = []

for row in newdata:
    cleaned_row = [item for item in row if item != 0]
    if cleaned_row:
        filtered_data.append(cleaned_row)

newdata = filtered_data
newdata[:10]
```

```python
from apyori import apriori

rules = list(apriori(newdata, min_support=0.00026, min_confidence=0.4, min_length=2, max_length=3, min_lift=1.3))
```

```python
sorted_rules = []

for record in rules:
    for rule in record.ordered_statistics:
        sorted_rules.append({
            'base': tuple(rule.items_base),
            'add': tuple(rule.items_add),
            'support': record.support,
            'confidence': rule.confidence,
            'lift': rule.lift
        })

sorted_rules = sorted(sorted_rules, key=lambda x: x['support'], reverse=True)

i = 1
for rule in sorted_rules[:10]:
    print(f"Rule #{i}\n {rule['base']} -> {rule['add']}")
    print(f"support: {rule['support']:5f}")
    print(f"confidence: {rule['confidence']:.2f}")
    print(f"lift: {rule['lift']:.2f}\n")
    i += 1
```
